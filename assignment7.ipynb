```python
import csv


while True:
    print("\n1.Add Contact Details")
    print("2.View Contacts")         
    print("3.Exit\n")       

    choice=int(input("Enter your choice : "))
    print(f"Your choice : {choice}\n")

    if choice==1:
        name=input("Enter your Name : ")
        mobile=input("Enter your Mobile Number : ")
        email=input("Enter your Email : ")

        lst=[[name,mobile,email]]

        with open("address_book.csv","a") as f:
            writer=csv.writer(f)
            for row in lst:
                writer.writerow(row)

    elif(choice==2):
        print("Contact Details : \n")
        with open("address_book.csv") as f:
            reader=csv.reader(f)
            for row in reader:
                print(row)

    elif(choice==3):
        print("Exited")
        break        

```
