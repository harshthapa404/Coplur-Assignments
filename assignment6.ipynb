```python
#Assignment 6 Question 2

#In this we have to add unique list and remove the repeated list items

def d_e(number):    #Here we created a function name d_e() and giving a list of numbers.
    
    result=[]       #This is a empty list

    for items in number:        #Here we are storing numbers into a items

        if items not in result:     #if items not present in a result than add that item to result.

            result.append(items)       #This will add items to a empty result that is not present in the list
            
    return result                 #this will return the value of result like 1,2,3 etc only the unique value.


number=[1,2,3,4,4,5,5,6,6,7,7,8,9]

print(d_e(number))  #here we call a function.
```

```python
#Assignment 6 Question 3

#In this we multiply the numbers that is present in a list using function.

def product(lst): #lst contain numbers like this [2,3,7,9,7].

    m=1

    for i in lst:    #lst values assign to the i only by one 
        
        m=m*i 
    
    return m         #This will print the product of the numbers.

lst=[2,3,7,8] 

product(lst)
```

```python
#Assignment 6 Question 4

#Finding the factorial of a number using function.

def factorial(n):

    m=1

    for i in range(1,n+1):
         
         m=m*i

    print(f"Factorial of a {n} is {m}")


n = int(input("Enter number: "))

factorial(n)
    
```

```python
#Assignment 6 Question 8

#Using function check whether the number is prime or not.

def p(number):

    if number<=1:      #checking if 1 is less than or equal to than print Number is not prime.
        print(f"{number} is not Prime Number")
    else:
        for i in range(2,number):

            if number%i==0:     #if it's remainder = 0 than it is not a prime number because prime number only divisible by 1 and itself.
                print(f"{number} is not Prime Number")
                break

        else:                                   #if it's remainder !=0 than it is a prime number.
            print(f"{number} is Prime Number")  

number=int(input("Enter number: "))
p(number)
```

```python
#Assignment 6 Question 9

#In this we have to count the upper and lower case of the string.

def u_l(s):

    u=0 
    l=0

    for i in s:

        if i.isupper():     #Only runs when the value of (i) is True. if it is run we store a value in u to check how many times loops run.
            u=u+1
            
        elif i.islower():
            l=l+1
    print(f"The String we provide : {s}\n")     
    print(f"Upper case in a string : {u}\n")
    print(f"Lower case in a String : {l}")

s=input("Enter any String : ")

u_l(s)
```
