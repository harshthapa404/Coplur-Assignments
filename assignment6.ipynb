```python
#Assignment 6 Question 2

#In this we have to add unique list and remove the repeated list items

def d_e(number):    #Here we created a function name d_e() and giving a list of numbers.
    
    result=[]       #This is a empty list

    for items in number:        #Here we are storing numbers into a items

        if items not in result:     #if items not present in a result than add that item to result.

            result.append(items)       #This will add items to a empty result that is not present in the list
            
    return result                 #this will return the value of result like 1,2,3 etc only the unique value.


number=[1,2,3,4,4,5,5,6,6,7,7,8,9]

print(d_e(number))  #here we call a function.
```

```python
#Assignment 6 Question 3

#In this we multiply the numbers that is present in a list using function.

def product(lst): #lst contain numbers like this [2,3,7,9,7].

    m=1

    for i in lst:    #lst values assign to the i only by one 
        m=m*i 
    
    return m         #This will print the product of the numbers.

lst=[2,3,7,9] 

product(lst)
```
