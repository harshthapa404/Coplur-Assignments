```python
#Assignment 3 Question 1

#Student Enter Name,Class and Marks of 5 Subjects
#And than this program will generate the marksheet of the Student with Marks of each subject,total marks,percentage and Grade of a Student.
name=input("Enter Student Name : ")

standard=input("Enter your class : ")

english=int(input("Enter your English Marks : "))

hindi=int(input("Enter your Hindi Marks : "))

mathematic=int(input("Enter your Mathematic Marks : "))

chemistry=int(input("Enter your Chemistry Marsk : "))

biology=int(input("Enter your Biology Marks : "))

totalmarks=english+hindi+mathematic+chemistry+biology

percentage=(totalmarks/500)*100

print(f'''Name : {name}
      
Class : {standard}

Subjects       Marks

English         {english}
Hindi           {hindi}
Mathematics     {mathematic}
Chemistry       {chemistry}
Biology         {biology}

Total Marks     {totalmarks}

Percentage      {percentage}
''')

if(percentage>=60):
    print(f"{name}'s overall Grade is A")

elif(percentage>=50) and (percentage<60 ):
    print(f"{name}'s overall Grade is B")

elif(percentage>=40) and (percentage<50):
    print(f"{name}'s overall Grade is C")

elif(percentage>=33) and (percentage<40):
    print(f"{name}'s overall Grade is D")

else:
    print("Grade           Fail")
```

```python
#Assignment 3 Question 2.

#User can generate table from 2 to 50.
#If user want to generate table of 51,52,53 and so on .than this will print number is not in a range.

n=int(input("Enter number to generate a Table: "))
print(f"Table of {n}")
for i in range(1,11):
    if (n<=50):
        print(f"{n} X {i} = {n*i}")

    else:
        print(f"{n} is not in range")
        break
```

```python
#Assignment 3 Question 3.

#In this program user will input a number and we will check if the number is Palindrome or not 
#if we reverse that number and it will generate the same number that we entered than it is a palindrome number otherwise not.
#For example: 12321,121,131,1441 etc are palindrome numbers
num=input("Enter number")

rev=(num[-1::-1]) #It will reverse the number you entered. start:stop:steps.

if (num==rev):
    print("It is a Palindrome Number:",num)
else:
    print("It is not a Palindrome Number:",num)
```
