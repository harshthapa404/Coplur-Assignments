```python
#Assignment 10 Question 1

#In this we have to create a Class Rectangle and have two instance attributes : Length and Width. 
#And we have to create two Method attributes : Area and Perimeter.
#And find the Area of the Rectangle.

class Rectangle:

    def __init__(self,length,width):        
        self.length=length          #This is a Instance attribute
        self.width=width 

    def area(self):                 
        print(f"Length : {self.length}")    #This is a Method attribute
        print(f"Width : {self.width}")
        print(f"The Area of the Rectangle is : {self.length*self.width}") 

    def perimeter(self):
        print(f"\nLength : {self.length}")
        print(f"Width : {self.width}")
        print(f"The Perimeter of the Rectangle is : {2*(self.length + self.width)}")


a=Rectangle(5,3)        #One Object is created of class Rectangle name (a) and values are passed 
a.area()                #Calling method (area) with the object (a) values 5,3.
p=Rectangle(5,3)
p.perimeter()

```

```python
#Assignment 10 Question 2

#In This we have to create a class Circle.
#And make a instance attribute with name Radius.
#And create three methods.
#That returns DIAMETER,ARE and CIRCUMFERENCE(perimeter).


class Circle:
    
    def __init__(self,radius):
        self.radius=radius      #This is a instance attribute

    def diameter(self):         #This is a method attribute
        print(f"Radius : {self.radius}")    
        print(f"The Diameter of a Circle : {2*self.radius} ")

    def area(self):                         #This is a method attribute
        print(f"\nRadius : {self.radius}")  
        print(f"The Area of a Circle : {(22/7*(self.radius**2)):.2f}")

    def circumference(self):                #This is a method attribute
        print(f"\nRadius : {self.radius}")
        print(f"The Circumference of a Circle : {(2*22/7*self.radius):.2f}")


d=Circle(4)
d.diameter()
a=Circle(2)
a.area()
c=Circle(5)
c.circumference()
```

```python
#Assignment 10 Question 3

#In this we have to create a class Product.
#And make Three Instance Attribute Product Name,Price and Quantity.
#And make one method total_value that return total stock value by multiplying the price of the product and the quantiy of the product.

class Product:
    
    def __init__(self,name,price,quantity):
        self.name=name
        self.price=price 
        self.quantity=quantity 


    def total_value(self):
        print(f"Product : {self.name}")
        print(f"\nPrice of a Product : {self.price}")
        print(f"\nQuantity of a Product : {self.quantity}")
        print(f"\nTotal Value of a Product : {self.price*self.quantity}")
        


p1=Product("Rice",350,20)
p1.total_value()
```

```python
#Assignment 10 Question 4

#In this we have to Create a class BankAccount.
#And we have to create two method : Deposit(amount) and withdrawn(amount)
#And if anyone want to withdraw the amount and that is less than the balance or equal to the balance than it will withdraw.
#Otherwise it will show   

class BankAccount:

    def __init__(self,balance=0):
        self.balance=balance
    
    def deposit_amount(self,amount):
        self.balance = self.balance + amount
        print(f"\nDeposit Amount : {amount}")
        print(f"\nCurrent Balance : {self.balance}")


    def withdraw_amount(self,amount):
        if amount <= self.balance:
            self.balance = self.balance - amount
            print(f"\nWithdrawn Amount : {amount}")
            print(f"\nCurrent Amount : {self.balance}")

        else: 
            print("\nInsufficient Balance Withdrawn denied")




add=BankAccount(5000)       #This is the balance amount that we have.

add.deposit_amount(500)     #This is the deposit amount that we deposited.

add.withdraw_amount(500)    #This is the withdrawn amount that i want to withdraw.

add.withdraw_amount(5500)  #This is the withdrawn amount that i want to withdraw.
```

```python
#Assignment 10 Question 5 

#In this we have to create a class User.
#And we create a method check_password and we have to pass the value in input_password.
#And than we have to validate the password with input_password.


class User:


    def __init__(self,username,password):
        self.username=username          #this is a instance attribute
        self.password=password


    def check_password(self,input_password):        #this is a method attribute
        print(f"User-Name : {self.username}\n")
        if (self.password == input_password):       #Here we are checking the password with input_password.
            print("Password : True")
        else:
            print("Password : False")

u1=User("Harsh","Helloworld")       #First we pass the user value and than password value.
u1.check_password("helloworld")     #Here we pass the value of input_password and than check with password.
```
