```python

#Assignment 13 Question 1.

#In this we have to create a class Temperature
#And create two methods : celsius to fahrenheit and celsius into kelvin.

class Temperature:
    def __init__(self,temp_in_celsius): #First we created a constructor that holds the value of celsius so i can use in both methods.
        self.temp_in_celsius=temp_in_celsius

    def to_fahrenheit(self):            #Converting celsius into fahrenheit
        print(f"\n {self.temp_in_celsius} Celsius into Fahrenheit : {(self.temp_in_celsius*1.8)+32} F")

    def to_kelvin(self):                #Converting celsius into kelvin
        print(f"\n {self.temp_in_celsius} Celsius into Kelvin : {self.temp_in_celsius + 273.15} K")


f1=Temperature(25)      #created a object

f1.to_fahrenheit()      #calling object one with to_fahrenheit method.This will convert the value into fahrenheit.

f1.to_kelvin()          #calling object one with to_kelvin method.This will convert the value into kelvin.
```

```python
#Assignment 13 Question 3 

#In This we have to create a class CoffeeMachine
#And check resources are available or not to make latte coffee.
#If it is available than print yes,Latte coffee is available and can also display the remaining resources of the Latte coffee.
#Otherwise print I apologize, but Latte is currently unavailable.

class CoffeeMachine:

    def __init__(self,water,coffee,milk):
        self.water=water
        self.coffee=coffee
        self.milk=milk
    

    
    def make_latte(self):
        if (self.water>=200 and self.coffee>=20 and self.milk>=150):
            print("Yes, Latte coffee is available")
            self.water=self.water-200
            self.coffee=self.coffee-20
            self.milk=self.milk-150
        else:
            print("I apologize, but Latte coffee is currently unavailable")
    
    

    def display_latte_ingredients(self):
        print("\nRemaining Resources")
        print(f"Water(ml) : {self.water}")
        print(f"Coffee(g) : {self.coffee}")
        print(f"Milk(ml) : {self.milk}")

coffee1=CoffeeMachine(300,30,200)   #Values of the water,coffee and milk.
coffee1.make_latte()
coffee1.display_latte_ingredients()
```

```python
class Vehicle:
    def __init__(self,name,max_speed):
        self.name=name
        self.max_speed=max_speed

    def display(self):
        print(f"Bus Name : {self.name}")
        print(f"Bus Max Speed : {self.max_speed}")

class Bus(Vehicle):
    def __init__(self,name,max_speed,seats_booked,seating_capacity=50):
        self.name=name
        self.max_speed=max_speed
        self.seats_booked=seats_booked
        self.seating_capacity=seating_capacity

    def reservation(self):
        if (self.seating_capacity>=self.seats_booked):
            print(f"\nBus Booked Seats : {self.seats_booked}")
            print(f"Bus Seats Capacity : {self.seating_capacity}")
            print(f"\nYou can book {self.seats_booked} Seats because Bus Capacity is : {self.seating_capacity}")
        else:
            print(f"\nBus Seats Capacity : {self.seating_capacity}")
            print(f"You cannot book {self.seats_booked} Seats because Bus Capacity is : {self.seating_capacity}")
        



obj1=Bus("Maha Laxmi",50,50)
obj1.display()
obj1.reservation()

```
