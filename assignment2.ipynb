```python
#Assignment 2 Question 1
#Print Student Name ,Class and Marks Entered by the Student and print total marks and percentage of the student
name=input("Enter Student Name : ")

standard=input("Enter your class : ")

english=int(input("Enter your English Marks : "))

hindi=int(input("Enter your Hindi Marks : "))

mathematic=int(input("Enter your Mathematic Marks : "))

chemistry=int(input("Enter your Chemistry Marsk : "))

biology=int(input("Enter your Biology Marks : "))

totalmarks=english+hindi+mathematic+chemistry+biology

print("Name : ",name,"\n\nClass : ",standard,"\n")

print("Subjects","Marks\n",sep="        ")

print("English",english,sep="          ")

print("Hindi",hindi,sep="            ",)

print("Mathematics",mathematic,sep="      ")

print("Chemistry",chemistry,sep="        ")

print("Biology",biology,"\n",sep="          ")

print("Total Marks",totalmarks,sep="      ")

print("\nPercentage",totalmarks/500*100,sep="      ")
```

```python
#Assignment 2 Question 2
#String Method in Python
a="Welcome to Python Program."
b="Offered by Coplur"
c=a + " "+b
print("1",c.lower(),"\n")#Convert string into lowercasea(Small letter).


print("2",c.upper(),"\n")#Convert string into uppercase(Capital letter).


print("3",c.center(100,"-"),"\n")#Pad the string with the specified character.


print("4",c.swapcase(),"\n")#Convert string into lowercase to uppercase and uppercase to lowercase.


print("5",c.casefold(),"\n")#Convert string into lower case.


print("6",c.title(),"\n")#Convert string all first word's alphabet into upper case.


print("7",c.capitalize(),"\n")#Convert string first alphabet into upper case.


print("8",c.count("to"),"\n")#Returns the number of occurrences of a substring in the string.


print("9",c.endswith("r"),"\n")#Returns True if it's end with the word/alphabet you provide otherwise False.


print("10",c.find("to"),"\n")#Returns the first word alphabet's index and also give alphabet index.


print("11",c.isalnum(),"\n")#Returns true if the string only contain alphabet and numbers otherwise False.


print("12",c.isdigit(),"\n")#Returns true if all characters in the string are digits otherwise False.


print("13",c.isnumeric(),"\n")#Returns true if all character in the string are numeric other False.


print("14",c.isspace(),"\n")#Returns true if all the characters in the string are whitespaces characters otherwise False.


print("15",c.replace("Welcome","Most Welcome"),"\n")#Replace the value from old to new one that you provide.

```
